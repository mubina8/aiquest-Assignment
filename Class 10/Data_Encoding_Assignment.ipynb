## **Label Encoder**

```python
df1.sex = le.fit_transform(df1[['sex']])
df1.smoker = le.fit_transform(df1[['smoker']])
df1.region = le.fit_transform(df1[['region']])
```

## **Label Encoding with Label Mapping**

```python
df2.sex.unique()
```

```python
df2.smoker.unique()
```

```python
df2.region.unique()
```

```python
Label_mapping_sex = {'female': 0, 'male' : 1}
Label_mapping_smoker = {'yes': 1, 'no' : 0}
Label_mapping_region = {'southwest': 0, 'southeast' : 1, 'northwest' : 2, 'northeast' : 3 }
```

```python
Label_mapping_sex.keys()
```

```python
Label_mapping_sex.values()
```

```python
df2.sex = [Label_mapping_sex[sex] for sex in df2.sex]
df2.smoker = [Label_mapping_smoker[smoker] for smoker in df2.smoker]
df2.region = [Label_mapping_region[region] for region in df2.region]
```

```python
df2.head()
```

# **Label Encoding with Replace Method**

```python
df3.head()
```

```python
df3['sex'] = df3['sex'].replace(Label_mapping_sex)
df3['smoker'] = df3['smoker'].replace(Label_mapping_smoker)
df3['region'] = df3['region'].replace(Label_mapping_region)
```

```python
df3.head()
```

# **One Hot Encoder**

```python
df4.head()
```

```python
dummy_sex = pd.get_dummies(df4['sex'], drop_first = True, prefix = 'sex', prefix_sep='_')
dummy_smoker = pd.get_dummies(df4['smoker'], drop_first = True, prefix = 'smoker', prefix_sep='_')
dummy_region = pd.get_dummies(df4['region'], drop_first = True, prefix = 'region', prefix_sep='_')
```

```python
dummy_region.head()
```

```python
new_df4 = pd.concat([df4,dummy_sex,dummy_smoker,dummy_region],axis=1)
```

```python
new_df4.head()
```

```python
new_df4.drop(['sex','smoker','region'],axis=1,inplace=True)
```

```python
new_df4.head()
```

# **Ordinal Encoding**

```python
df5.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal = OrdinalEncoder(categories = [['female','male']])
ordinal_smoker = OrdinalEncoder(categories = [['yes','no']])
ordinal_region = OrdinalEncoder(categories = [['southwest','southeast','northwest','northeast']])
```

```python
df5.sex = ordinal.fit_transform(df5[['sex']])
df5.smoker = ordinal_smoker.fit_transform(df5[['smoker']])
df5.region = ordinal_region.fit_transform(df5[['region']])
```
