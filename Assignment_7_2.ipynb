```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Data Science project/AI Quest/weight-height.csv')
```

```python
data.head()
```

```python
data_copy = data.copy()
data_copy2 = data.copy()
```

```python
data_copy['Gender'].unique()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

**Scaling the Height using Standard Scaler**

```python
data_copy['Height'] = scaler.fit_transform(data_copy[['Height']])
```

**Converting Gender Column into numerical value using label encoder**
**Male = 1 & Female = 0**

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
data_copy['Gender']= label_encoder.fit_transform(data_copy['Gender'])
```

```python
data_copy.head()
```

```python
X = data_copy[['Gender','Height']]
```

```python
y = data_copy['Weight']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=.30, random_state=42)
```

```python
xtrain.head()
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(xtrain, ytrain)
```

```python
reg.predict(xtest)
```

```python
reg.score(xtest, ytest)
```

```python
reg.score(xtrain, ytrain)
```

```python
data['predicted_Weight_by_LR'] = reg.predict(X)
```

```python
data.head()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(data['Weight'], data['predicted_Weight_by_LR'])
mse
```

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=5)
neigh.fit(xtrain, ytrain)
```

```python
neigh.predict(xtest)
```

```python
neigh.score(xtest, ytest)
```

```python
neigh.score(xtrain, ytrain)
```

```python
data['predicted_Weight_by_KNN'] = neigh.predict(X)
```

```python
data.head()
```

```python
from sklearn.metrics import mean_squared_error
mse_knn = mean_squared_error(data['Weight'], data['predicted_Weight_by_KNN'])
mse_knn
```
