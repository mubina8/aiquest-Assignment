```python
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/Data Science project/AI Quest/Age_Income_Dataset.csv")
```

```python
data.head()
```

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=3)
```

```python
neigh.fit(data[['age']], data['Income(k)'])
```

**Validating Manual Calculations to Knn algorithm Calculations**

```python
neigh.predict([[22]])
```

**Now Comparing Linear Regression result to KNN algorithm result**

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(data[['age']], data['Income(k)'])
```

```python
reg.predict([[22]])
```

**Comparing the Predicted outcomes**

```python
data['KNN_Prediction_for_Income'] = neigh.predict(data[['age']])
data['LinearReg_Prediction_for_Income'] = reg.predict(data[['age']])
```
